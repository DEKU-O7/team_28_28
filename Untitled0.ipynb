```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the standard value is in nM, convert to Molar (M)
# filter out any zero or negative values as log10 won't work on those
filter_df['Molar'] = filter_df['Standard Value'] * 1e-9  # converting nM to M
filter_df = filter_df[filter_df['Molar'] > 0]  # filter out zero or negative values

# Calculate pIC50
filter_df['pIC50'] = -np.log10(filter_df['Molar'])

# Generate the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=filter_df, y='pIC50')
plt.title('Box plot of pIC50 values')
plt.ylabel('pIC50')
plt.show()
```
